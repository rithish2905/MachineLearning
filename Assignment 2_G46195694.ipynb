### (2)	Simply use k-means clustering to cluster all bike docking stations for k value from 2 to 50 and plot the curve for within-cluster sum of squares (WSS) vs k. Discuss your choice of the number of crew, say $k^*$, based on the curve. (We will use this $k^*$ value throughout the following questions.) [1 pt]

Answer:

```python
# K-means clustering
from sklearn.cluster import KMeans

cluster_range = range(2,50)
cluster_wss = []

for num_cluster in cluster_range:
    clusters = KMeans(num_cluster)
    clusters.fit(sta)
    cluster_wss.append(clusters.inertia_)
    
plt.xlabel('# Clusters')
plt.ylabel('WSS')
plt.plot(cluster_range, cluster_wss, marker = 'o')
plt.show()
```

### (3) Given your choice of optimal k*. Visualize the clusters of stations from k-means clustering. [1 pt]

Answer:

```python
# K-means clustering
from sklearn.cluster import KMeans
K=10 # you may choose your own K
kmeans = KMeans(n_clusters=K, random_state=0)

# cluster membership for each data point
clusters=kmeans.fit_predict(sta)
# get centroids
centers = kmeans.cluster_centers_

# plot the clusters and centroids
plt.figure(figsize=(9,12))
plt.scatter(sta['Longitude'], sta['Latitude'], c=clusters)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100);
plt.title("K-means Clustering")
plt.show()
```

### (4) As discussed in class, the objective function in k-means clustering may not be appropriate. If we want to minimize the total distance, the crew visits bike stations from the center. Instead, we will use k-median clustering that considers Manhattan distance.  Please find the $k^*$ clusters using k-median clustering and visualize them. [1 pt]

Answer:

```python
pip install pyclustering
```

```python
# install package pyclustering (to the right directory...carefully)
from pyclustering.cluster.kmedians import kmedians
from pyclustering.cluster import cluster_visualizer
from pyclustering.cluster.center_initializer import random_center_initializer

K=10
# Create instance of K-Medians algorithm.
initial_medians = random_center_initializer(sta.values,K).initialize() # randomly initialize centers
kmedians_instance = kmedians(sta.values.tolist(),initial_medians)

# Run cluster analysis and obtain results.
kmedians_instance.process()
cluster_instances = kmedians_instance.get_clusters() # index of data points in each cluster
centers = np.array(kmedians_instance.get_medians())

clusters=np.zeros(len(sta.index)) # initialize the membership of each point
for k in range(K):
    clusters[cluster_instances[k]]=k # label cluster membership for each point

plt.figure(figsize=(9,12))
plt.scatter(sta['Longitude'], sta['Latitude'], c=clusters)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100);
plt.title("K-medians Clustering")
plt.show()
```

### (5) If BlueBike wants to setup a depot for each crew member at a selected bike station, we can conduct k-medoids clustering to identify the centers and recommend them as the depots. Please identify the $k^*$ clusters and their centers from k-medoids clustering and visualize them. [1 pt]

Answer:

```python
from pyclustering.cluster.kmedoids import kmedoids

# Set random initial medoids.
# Set random initial medoids for 12 clusters.
initial_medoids = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250]
# provide initial medoids using indexes; 

# Create instance of K-Medoids algorithm.
kmedoids_instance = kmedoids(sta.values, initial_medoids)
kmedoids_instance.process()

# Run cluster analysis and obtain results.
cluster_instances = kmedoids_instance.get_clusters() # index of data points in each cluster
center_index = kmedoids_instance.get_medoids()
centers=sta.values[center_index,:]

clusters=np.zeros(len(sta.index)) # initialize the membership of each point
for k in range(K):
    clusters[cluster_instances[k]]=k # label cluster membership for each point

plt.figure(figsize=(9,12))
plt.scatter(sta['Longitude'], sta['Latitude'], c=clusters)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100);
plt.title("K-medoids Clustering")
plt.show()
```

### (6) Conduct hierarchical clustering to generate the $k^*$ clusters and visualize them. [1 pt]

Answer:

```python
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(sta, 'ward') # use ward linkage
fig = plt.figure(figsize=(15, 10)) # plot the dendrogram
dn = dendrogram(Z)
```

```python
from scipy.cluster.hierarchy import ward, fcluster
clusters = fcluster(Z, t=0.1, criterion='distance') # get clusters by cutting the dendrogram at given distance at "t=0.1"
# print(clusters)

plt.figure(figsize=(9,12))
plt.scatter(sta['Longitude'], sta['Latitude'], c=clusters, cmap='hsv')
plt.title("Hierarchical Clustering from scipy")
plt.show()
```
